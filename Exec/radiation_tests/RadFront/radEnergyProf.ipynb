```python
from matplotlib import pyplot as plt

import yt
import numpy as np
```

```python
ds = yt.load("plt_01540")

ray = ds.ortho_ray(0,(0,0))
ray_sort = np.argsort(ray["x"])

x = np.array(ray['x'][ray_sort])


plt.plot(x, np.array(ray['rad'][ray_sort]))

plt.show()
```

```python
#For the 2d setup
ds = yt.load("plt_02309")

#making the profile
ray = ds.ortho_ray(1,(0,0))
ray_sort = np.argsort(ray["y"])

z = np.array(ray['y'][ray_sort])

plt.figure(3)
plt.plot(z, np.array(ray['rad'][ray_sort]))
plt.show()
```
